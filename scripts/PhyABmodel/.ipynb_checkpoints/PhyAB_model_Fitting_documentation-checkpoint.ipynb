A trick has to be performed with the cost so that we can use sloppycell. SloppyCell will required to introduce a time series 
however we are intersted in the ration between two quantities. I will intruce a prior that will allow me to calculate that ration
Developing this parts will required to use the terminal. This can fixed by performing a Nohup and then running in background the process

I am choosing to use sloppycell becuase I will be able to get uncertaintanties for the model parameters. 

SloppyCell requiers data for running, becuase it is assuming that it will infer parameters form time series. Instead Im am going to use the reation between two quantities. I build a data dictionary which contains the variable for which sloppycell has to generate predictions. However becase this is just dummy data, I ncrease the uncertainties dramatically so that it is not considered.

, however I will not feed it with time series

```python
data={}
for idx, i in enumerate(Am.columns):
    if 'W' in i:
        next
    else:
        d = {}
        d['Am'] = {80:(Am.iloc[0,idx],1e10), 190:(Am.iloc[0,idx],1e10)}
        data[i] = d
data
```

Exporting models into SloppyCell

```python
networks={}
networks['Ler_R'] = sc.IO.from_SBML_file('SloppyCell_model/phyAB_interaction_WT.xml', 'Ler_R',duplicate_rxn_params=True)
networks['Ler_R'].add_event('on_R','gt(time,100)',{'R':1.})
networks['Ler_R'].add_event('on_FR','gt(time,100)',{'FR':0})

networks['Ler_FR'] = sc.IO.from_SBML_file('SloppyCell_model/phyAB_interaction_WT.xml', 'Ler_FR',duplicate_rxn_params=True)
networks['Ler_FR'].add_event('on_R','gt(time,100)',{'R':0})
networks['Ler_FR'].add_event('on_FR','gt(time,100)',{'FR':1})

networks['phyB_5_FR'] = sc.IO.from_SBML_file('SloppyCell_model/phyAB_interaction_phyBm.xml', 'phyB_5_FR',duplicate_rxn_params=True)
networks['phyB_5_FR'].add_event('on_R','gt(time,100)',{'R':0})
networks['phyB_5_FR'].add_event('on_FR','gt(time,100)',{'FR':1})

networks['phyB_5_R'] = sc.IO.from_SBML_file('SloppyCell_model/phyAB_interaction_phyBm.xml', 'phyB_5_R',duplicate_rxn_params=True)
networks['phyB_5_R'].add_event('on_R','gt(time,100)',{'R':1.})
networks['phyB_5_R'].add_event('on_FR','gt(time,100)',{'FR':0})


networks['RLD_R'] = sc.IO.from_SBML_file('SloppyCell_model/phyAB_interaction_WT.xml', 'RLD_R',duplicate_rxn_params=True)
networks['RLD_R'].add_event('on_R','gt(time,100)',{'R':1.})
networks['RLD_R'].add_event('on_FR','gt(time,100)',{'FR':0})

networks['RLD_FR'] = sc.IO.from_SBML_file('SloppyCell_model/phyAB_interaction_WT.xml', 'RLD_FR',duplicate_rxn_params=True)
networks['RLD_FR'].add_event('on_R','gt(time,100)',{'R':0})
networks['RLD_FR'].add_event('on_FR','gt(time,100)',{'FR':1})

networks['phyA_101_FR'] = sc.IO.from_SBML_file('SloppyCell_model/phyAB_interaction_phyAm.xml', 'phyA_101_FR',duplicate_rxn_params=True)
networks['phyA_101_FR'].add_event('on_R','gt(time,100)',{'R':0})
networks['phyA_101_FR'].add_event('on_FR','gt(time,100)',{'FR':1})


networks['phyA_101_R'] = sc.IO.from_SBML_file('SloppyCell_model/phyAB_interaction_phyAm.xml', 'phyA_101_R',duplicate_rxn_params=True)
networks['phyA_101_R'].add_event('on_R','gt(time,100)',{'R':1})
networks['phyA_101_R'].add_event('on_FR','gt(time,100)',{'FR':0})


```

Then the cost function experimental data has to be constructed. 

```python
expt = sc.Experiment('Const_light')
expt.set_data(data)
#expt.set_fixed_sf({'Am':1})
```

```python
m = sc.Model([expt],networks.items())

m.AddResidual(sc.Residuals.Ratio('Ratio_RLD_R_Am','RLD_R','Am',80,190,40.894,4.011))
m.AddResidual(sc.Residuals.Ratio('Ratio_RLD_R_Am','RLD_FR','Am',80,190,33.023,5.401))

m.AddResidual(sc.Residuals.Ratio('Ratio_phyA_101_R_Am','phyA_101_R','Am',80,190,49.689,8.796))
m.AddResidual(sc.Residuals.Ratio('Ratio_phyA_101_FR_Am','phyA_101_FR','Am',80,190,10.0304,4.938))

m.AddResidual(sc.Residuals.Ratio('Ratio_Ler_R_Am','Ler_R','Am',80,190,43.054,3.086))
m.AddResidual(sc.Residuals.Ratio('Ratio_Ler_FR_Am','Ler_FR','Am',80,190,24.536,9.722))

m.AddResidual(sc.Residuals.Ratio('Ratio_phyB_5_R_Am','phyB_5_R','Am',80,190,90.583,10.802))
m.AddResidual(sc.Residuals.Ratio('Ratio_phyB_5_FR_Am','phyB_5_FR','Am',80,190,29.628,9.568))

params = m.get_params()
print m.cost(params)
```

I need to modify SloppyCell that the minimisation function works with finite difference given that I am using a new type of residual

```python
Model_params = sc.Optimization.fmin_lm_log_params_fd(m,params)
```

```python
#sc.Utility.save(params,'params.bp')
Model_params = sc.Utility.load('params.bp')
```

Then we plot the model vs the fitted data. 

```python
phyABte = te.loadAntimonyModel('SloppyCell_model/phyAB_interaction')
for p in Model_params.keys():
    phyABte.setValue(p,Model_params.getByKey(p))

phyABte.setValue('R_levels',1)
phyABte.setValue('muta',0)
phyABte.setValue('mutb',0)
phyABte.setValue('FR_levels',0)
    
phyABteres = phyABte.simulate(0,200,1000)
#bar(1,(phyABteres['[Am]']/max(phyABteres['[Am]'])))
plot(phyABteres['time'],phyABteres['[Bm]'])
```
